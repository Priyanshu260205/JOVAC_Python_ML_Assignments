```python
import numpy as np
import pandas as pd 
import seaborn as sns
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
```

```python
# 1. Load dataset into a DataFrame for easier analysis
housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df['MedHouseVal'] = housing.target
```

```python
# 2. Correlation matrix
corr_matrix = df.corr()
```

```python
# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 3. Find features with strongest relationship to target
target_corr = corr_matrix['MedHouseVal'].drop('MedHouseVal')
sorted_corr = target_corr.sort_values(ascending=False)
```

```python
print("\nCorrelation of each feature with MedHouseVal:\n")
print(sorted_corr)
```

```python
strongest_feature = sorted_corr.index[0]
print(f"\nFeature most strongly correlated with target: {strongest_feature}")
```
